# 1 / 4.

#### Split my data into a training and test set and create scatterplots, histograms, and a descriptive statistics table of the variables of interest.

```python
from sklearn.model_selection import train_test_split
```

```python
y = df_processed["salary"]
feature_cols = ["yearsworked","yearsrank","executive"]

X_1 = df_processed[feature_cols]
X_2 = df_processed.drop(["executive","yearsrank","junior employee","salary"], axis=1)
X_3 = df_processed.drop(["executive","marketing","junior employee","yearsabs",'salary'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_3, y, test_size=0.20, random_state=42)
```

```python
### STATSMODELS ###

X_train =  sm.add_constant(X_train)
X_test =  sm.add_constant(X_test)


# create a fitted model with all three features
model = sm.OLS(y_train, X_train).fit()

predictions = model.predict(X_train)
```

```python
model.save('model3_orig.pickle')
```

```python
model.summary()
```

```python
# print the coefficients
coef = model.params
coef
```

```python
# the coefficients represent the change in the output Y due to a change of one unit, by every unit yearsworked will increase by 2172
# yearsranked by 763 and executive by 8338.
# Yearsworked And position seem to play major roles in the salary
```

```python
pred_df = pd.DataFrame(predictions, columns=["predictions"])
pred_df
```

```python
std_residuals = statsmodels.regression.linear_model.RegressionResults.resid(model)
```

```python
fittedvalues = sm.regression.linear_model.RegressionResults.fittedvalues(model)
```

# 7

```python
# create a fitted model on train
model_train = sm.OLS(y_train, X_train).fit()
model_train.summary()
```

```python
# create a fitted model on test
model_test = sm.OLS(y_test, X_test).fit()
model_test.summary()
```

```python
# THE MODEL SEEMS TO BE PERFORMING BETTER ON THE TRAIN SET THAN THE TEST SET, INDICATING THERE MIGHT BE A CHANCE OF OVERFITTING.
```

```python
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
train_predict
```

```python
test_predict
```

```python
from sklearn import metrics
print('TRAIN RMSE:', np.sqrt(metrics.mean_squared_error(y_train, train_predict)))
print('TEST RMSE:', np.sqrt(metrics.mean_squared_error(y_test, test_predict)))
```

```python
# THE ISN'T MUCH OF A DIFFERECE IN RMSE BETWEEN THE TRAIN AND TEST SET ALSO SUGGESTING SOME OVERFITTING
```
